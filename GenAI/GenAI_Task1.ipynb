## Multi-Input Prompt System

```python
# Task 2: Multi-Input Prompt Template
multi_input_template = """
Explain the topic: {topic}
Target Audience: {audience}
Desired Tone: {tone}

Please provide a concise explanation based on these parameters.
"""

# Initialize the template with multiple variables
multi_prompt = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template=multi_input_template
)

# Test Case: Python for kids in a fun tone
formatted_prompt = multi_prompt.format(
    topic="Python Programming",
    audience="10-year-old kids",
    tone="fun and energetic"
)

print(formatted_prompt)
```

## Prompt Variations Engine

```python
# Task 3: Prompt Variations Engine
# Define a dictionary of different prompt styles
variations = {
    "Teaching": "Explain the concept of {topic} clearly with a step-by-step example.",
    "Interview": "Generate 3 challenging interview questions regarding {topic}.",
    "Storytelling": "Tell a short creative story that explains how {topic} works in the real world."
}

target_topic = "Machine Learning"

print(f"Generating variations for: {target_topic}\n" + "="*40)

# Loop through the variations and format each one
for style, template_str in variations.items():
    variation_prompt = PromptTemplate(input_variables=["topic"], template=template_str)

    print(f"Style: {style}")
    print(variation_prompt.format(topic=target_topic))
    print("-" * 20)
```

## ChatPromptTemplate System

```python
from langchain_core.prompts import ChatPromptTemplate

# Task 4: Chat System with Roles
chat_template = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful {role} who explains complex topics simply."),
    ("human", "Can you tell me about {topic}?")
])

# Example 1: Explaining to a student as a Teacher
messages1 = chat_template.format_messages(role="Science Teacher", topic="Photosynthesis")
print("--- Variation 1 (Teacher) ---")
for msg in messages1:
    print(f"{msg.type.capitalize()}: {msg.content}")

print("\n" + "="*40 + "\n")

# Example 2: Explaining to a developer as a Senior Engineer
messages2 = chat_template.format_messages(role="Senior Software Engineer", topic="API Gateways")
print("--- Variation 2 (Engineer) ---")
for msg in messages2:
    print(f"{msg.type.capitalize()}: {msg.content}")
```

## Input Validation Layer

```python
# Task 5: Input Validation Layer
def validate_inputs(audience, tone):
    # Define allowed lists
    valid_audiences = ["beginner", "intermediate", "expert"]
    valid_tones = ["formal", "casual", "fun"]

    # Check audience - default to beginner if invalid
    if audience.lower() not in valid_audiences:
        print(f"Warning: '{audience}' is not a valid audience. Defaulting to 'beginner'.")
        final_audience = "beginner"
    else:
        final_audience = audience.lower()

    # Check tone - default to formal if invalid
    if tone.lower() not in valid_tones:
        print(f"Warning: '{tone}' is not a valid tone. Defaulting to 'formal'.")
        final_tone = "formal"
    else:
        final_tone = tone.lower()

    return final_audience, final_tone

# Test Case: User provides an invalid audience "pro" and invalid tone "angry"
input_audience = "pro"
input_tone = "angry"

clean_audience, clean_tone = validate_inputs(input_audience, input_tone)

# Now we format the prompt using our "cleaned" inputs
validation_template = "Explain Python for a {audience} audience in a {tone} tone."
val_prompt = PromptTemplate(input_variables=["audience", "tone"], template=validation_template)

print("\nFinal Formatted Prompt:")
print(val_prompt.format(audience=clean_audience, tone=clean_tone))
```

## Prompt Generator App

```python
# Task 6: Master Prompt Generator Function
def generate_prompt(topic, audience, tone, style):
    # 1. Validate inputs (from Task 5)
    clean_audience, clean_tone = validate_inputs(audience, tone)

    # 2. Define the variation templates (from Task 3)
    styles = {
        "Teaching": "Explain {topic} to a {audience} in a {tone} way using a step-by-step guide.",
        "Interview": "As a {tone} interviewer, ask 3 questions about {topic} for an {audience} level.",
        "Storytelling": "Write a {tone} story about {topic} suitable for a {audience} audience."
    }

    # 3. Select the template based on style
    selected_template = styles.get(style, styles["Teaching"]) # Default to Teaching

    # 4. Create and format the PromptTemplate
    master_prompt = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=selected_template
    )

    return master_prompt.format(topic=topic, audience=clean_audience, tone=clean_tone)

# --- Test the Master System ---
final_output = generate_prompt(
    topic="Cloud Computing",
    audience="Beginner",
    tone="Fun",
    style="Storytelling"
)

print("--- FINAL GENERATED PROMPT ---")
print(final_output)
```

## Template Reusability Test

```python
# Task 7: Template Reusability Test
# 1. Define ONE reusable template
reusable_template = "Draft a {tone} {style} about {topic} for an {audience} audience."
prompt_obj = PromptTemplate(input_variables=["tone", "style", "topic", "audience"], template=reusable_template)

# 2. Define 5 different input scenarios
test_scenarios = [
    {"topic": "Blockchain", "audience": "beginner", "tone": "casual", "style": "blog post"},
    {"topic": "Cybersecurity", "audience": "expert", "tone": "formal", "style": "technical report"},
    {"topic": "Artificial Intelligence", "audience": "kids", "tone": "fun", "style": "short story"},
    {"topic": "Data Science", "audience": "intermediate", "tone": "professional", "style": "email newsletter"},
    {"topic": "Internet of Things", "audience": "expert", "tone": "urgent", "style": "security alert"}
]

print("--- Running Reusability Test (5 Scenarios) ---")
print("="*50)

# 3. Loop through scenarios and generate outputs
for i, scenario in enumerate(test_scenarios, 1):
    output = prompt_obj.format(**scenario)
    print(f"Scenario {i}:")
    print(output)
    print("-" * 30)
```
