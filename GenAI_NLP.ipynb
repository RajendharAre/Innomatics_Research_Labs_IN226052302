```python
import re
from collections import Counter

def preprocess_text(text):
    if not text or not isinstance(text, str):
        return []

    # 1. Convert to lowercase
    text = text.lower()

    # 2. Remove URLs and Emails
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)

    # 3. Handle repeated characters (e.g., "soooo" -> "so")
    # This regex looks for 3 or more repetitions and reduces to 1 or 2
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # 4. Remove numbers
    text = re.sub(r'\d+', '', text)

    # 5. Remove special characters/emojis (keeping basic punctuation for now)
    text = re.sub(r'[^\w\s]', '', text)

    # 6. Tokenize and Remove extra spaces
    tokens = text.split()

    # 7. Remove short tokens (length <= 2) except "no" and "not"
    exceptions = {"no", "not"}
    clean_tokens = [t for t in tokens if len(t) > 2 or t in exceptions]

    return clean_tokens

def full_pipeline(text_list):
    results = {
        "original": [],
        "tokens": [],
        "clean_sentences": []
    }

    for text in text_list:
        tokens = preprocess_text(text)
        results["original"].append(text)
        results["tokens"].append(tokens)
        results["clean_sentences"].append(" ".join(tokens))

    return results

# --- Task 3 & 4: Stress Testing & Analytics ---
sample_inputs = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

pipeline_output = full_pipeline(sample_inputs)

print(f"{'Original Text':<40} | {'Cleaned Sentence'}")
print("-" * 80)

all_tokens = []
for i in range(len(sample_inputs)):
    tokens = pipeline_output["tokens"][i]
    clean_sent = pipeline_output["clean_sentences"][i]
    all_tokens.extend(tokens)

    # Task 4: Analytics
    num_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_len = sum(len(t) for t in tokens) / num_tokens if num_tokens > 0 else 0

    print(f"{sample_inputs[i]:<40} | {clean_sent}")
    print(f"   -> Tokens: {num_tokens}, Unique: {unique_tokens}, Avg Len: {avg_len:.2f}\n")

# --- Task 5: Frequency Analysis ---
counts = Counter(all_tokens)
print("--- Frequency Analysis ---")
print("Top 10 Most Frequent:", counts.most_common(10))
print("Top 5 Least Frequent:", counts.most_common()[:-6:-1])
```
